```python
from pycaret.classification import *
import pandas as pd

# Carregar o dataset
data = pd.read_csv("/workspaces/auto-ml/metacritic_games.csv")

# Remover linhas com valores ausentes (NaN)
data = data.dropna()

# Definir um limiar para classificar jogos como "bons" ou "ruins"
threshold = 80
data['bom_jogo'] = (data['metascore'] >= threshold).astype(int)

# Selecionar as colunas relevantes
features = ['positive_critics', 'neutral_critics', 'negative_critics']
target = 'bom_jogo'

# Inicializar o ambiente PyCaret
clf = setup(data, target=target, 
           train_size=0.8,  # 80% para treino, 20% para teste
           session_id=123)  # Semente aleatória para reprodutibilidade

# Comparar modelos (opcional)
compare_models()

# Treinar o modelo escolhido (ex: LogisticRegression)
model = create_model('lr')

# Avaliar o modelo
evaluate_model(model)

# Salvar o modelo (opcional)
save_model(model, 'modelo_classificacao')

# Carregar o modelo (opcional)
loaded_model = load_model('modelo_classificacao')

# Fazer previsões com o modelo carregado (opcional)
predictions = predict_model(loaded_model, data=data)

# Visualizar as previsões
print(predictions)
```

```python
from autogluon.tabular import TabularPredictor

import pandas as pd

# Carregar o dataset
data = pd.read_csv("/workspaces/auto-ml/metacritic_games.csv")

# Remover linhas com valores ausentes (NaN)
data = data.dropna()
features = ['positive_critics', 'neutral_critics', 'negative_critics']
target = 'metascore'

# Inicializar o ambiente AutoGluon
predictor = TabularPredictor(label=target, problem_type='regression')

# Treinar o modelo com os dados
predictor = predictor.fit(data, presets='best_quality')

# Fazer previsões com o modelo treinado
predictions = predictor.predict(data)

# Avaliar o modelo
print(predictor.leaderboard(data))

# Salvar o modelo (opcional)
predictor.save("modelo_regressao")

# Carregar o modelo (opcional)
predictor = TabularPredictor.load("modelo_regressao")

# Fazer previsões com o modelo carregado (opcional)
predictions = predictor.predict(data)

# Visualizar as previsões
print(predictions)
```
